```python
def validate_user(name, email, password):
    """Check if the provided user inputs are valid. 
    
    Args:
        name (str): The user's name.
        email (str): The user's email.
        password (str): The user's password.

    Returns:
        bool: True if all inputs are valid. 

    Raises: 
        ValueError: If any input is invalid. 
    """
    if not validate_name(name): 
        raise ValueError("Please enter a name greater than two characters and is a string data type.")
    if not validate_email(email): 
        raise ValueError("Please enter an email address with a username greater than 1 character, an '@' symbol, and an allowed domain that is in the `top_level_domains` variable.")
    if not validate_password(password):
        raise ValueError("Please enter a password with a capital letter, a number between 0-9 and greater than 8 characters.")
 
    return True
```

```python
def register_user(name, email, password):
    """Register valid user input.
    
    Args:
        name (str): The user's name.
        email (str): The user's email.
        password (str): The user's password.

    Returns:
        dict: A dictionary containing the user's registration details if valid.
        bool: False if the input is invalid. 
    """
    try:
        validate_user(name, email, password)
        return {"name": name, "email": email, "password": password}
    except ValueError:
        return False
```
